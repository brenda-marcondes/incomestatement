```python
#SOMENTE PARA AÇÕES DA BOLSA NORTE AMERICANA
import pandas as pd
import numpy as np
from numpy import int64
from xml.dom.minidom import Attr
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from time import sleep
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
```

```python
#Dados Yahoo Finance #Do not change

stock = 'AAPL'

url = 'https://finance.yahoo.com/quote/' + stock + '/financials?p=' + stock

google = webdriver.Chrome()

google.get(url)

google.maximize_window()

print(url)
```

```python
#Finding data script

page_content =  google.page_source
site = BeautifulSoup(page_content, 'html.parser')

#Finding linhas with a couple of elements "div" inside
lines = site.findAll('div', attrs={'data-test' : 'fin-row'}) #Conjunto

#print(lines[2].prettify())

all_columns = []

for line in lines:
    columns = line.findAll('div')
    
    column_text = []
    for column in columns:
        column = column.text
        column_text.append(column)
    
    all_columns.append(column_text)
    
    #columns = square[6].text
    #print(square)
    
    
print(all_columns[7])
    
    
```

```python
#Plot data into a Dataframe
#Setting a name 

a = len(all_columns)
b = 0

all_dataframes_transposed = []

while a > b:

    df = pd.DataFrame(all_columns[b])

    df_transposto = df.T
    
    all_dataframes_transposed.append(df_transposto)
    
    b = b+1

#Cocatenating the columns

concat_df = pd.concat(all_dataframes_transposed, ignore_index=True)

#Droping the desExcluir as primeiras desnecessárias para concatenar com o tútulo

df = concat_df #Changind the name to the original dataframe

remove = [0,1,3,9]

df = concat_df.drop(df.columns[remove], axis=1)

df.head()
```

```python
#Findins the data time of the columns

title = site.find('div', attrs={'class' : 'D(tbhg)'})

for element in title:
    new = title.findAll('span')
    
dates = []

for element_text in new: # Convert into a text
    date = element_text.text
    dates.append(date)

#Turning the list of elements in title

df.columns = dates

df = df.set_index('Breakdown')

```
